```python
#plot GTWR result

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv("coefficient_cases.csv")
df.head()
```

```python
x = df.TIME
y1 = df.Brunei_Coef
y2 = df.Brunei_Cases/1000
y3 = df.Indonesia_Coef
y4 = df.Indonesia_Cases/1000
y5 = df.Malaysia_Coef
y6 = df.Malaysia_Cases/1000
y7 = df.Philippines_Coef
y8 = df.Philippines_Cases/1000
y9 = df.Singapore_Coef
y10 = df.Singapore_Cases/1000
y11 = df.Thailand_Coef
y12 = df.Thailand_Cases/1000
y13 = df.Vietnam_Coef
y14 = df.Vietnam_Cases/1000
```

```python
df['Brunei_Cases'] = df['Brunei_Cases'].apply(lambda x: format(int(x), ','))
df['Indonesia_Cases'] = df['Indonesia_Cases'].apply(lambda x: format(int(x), ','))
df['Malaysia_Cases'] = df['Malaysia_Cases'].apply(lambda x: format(int(x), ','))
df['Philippines_Cases'] = df['Philippines_Cases'].apply(lambda x: format(int(x), ','))
df['Singapore_Cases'] = df['Singapore_Cases'].apply(lambda x: format(int(x), ','))
df['Thailand_Cases'] = df['Thailand_Cases'].apply(lambda x: format(int(x), ','))
df['Vietnam_Cases'] = df['Vietnam_Cases'].apply(lambda x: format(int(x), ','))

x = df.TIME
y1 = df.Brunei_Coef
y2 = df.Brunei_Cases
y3 = df.Indonesia_Coef
y4 = df.Indonesia_Cases
y5 = df.Malaysia_Coef
y6 = df.Malaysia_Cases
y7 = df.Philippines_Coef
y8 = df.Philippines_Cases
y9 = df.Singapore_Coef
y10 = df.Singapore_Cases
y11 = df.Thailand_Coef
y12 = df.Thailand_Cases
y13 = df.Vietnam_Coef
y14 = df.Vietnam_Cases
```

```python
#define fig size
fig = plt.figure(figsize=(30,20))
ax1 = fig.add_subplot(111,facecolor='whitesmoke')

#set x and y
plt.xticks(rotation=0,size=45)
plt.yticks(size = 40)
plt.xlabel('Week',fontsize=60)
plt.style.use('seaborn')

#plot coefficient
line1 = ax1.plot(x,y1,linewidth=5,c='#77C4D1',label='coefficient')
ax1.set_ylim(0,1) #set y axis from 0 to 1
ax1.set_xlim(0.5,30.5)
ax1.set_ylabel('Coefficient of MOVE',fontsize=60)

#set events
ax1.vlines(10, 0,0.95,colors = '#FF7F75',ls='dashed',linewidth=3)#set when happened
ax1.annotate('COVID-19 Restriction',xy=(10,0.12),xytext=(11,0.15),fontsize=35,
            arrowprops=dict(arrowstyle='->', color='salmon',linewidth=3)) #set events introduction

ax1.vlines(19, 0,0.95,colors = '#FF7F75',ls='dashed',linewidth=3)
ax1.annotate('Restriction tightened',xy=(19,0.69),xytext=(12.5,0.69),fontsize=35, 
             arrowprops=dict(arrowstyle='->', color='salmon',linewidth=3))

ax1.vlines(29, 0,0.85,colors = '#74C69D',ls='dashed',linewidth=3)
ax1.annotate('''Transitted to 
Endemic phase''',xy=(29,0.02),xytext=(23.5,0.25),fontsize=35, 
             arrowprops=dict(arrowstyle='->', color='salmon',linewidth=3))

#plot cases
ax2=ax1.twinx()#ensure in the same fig
line2 = ax2.plot(x,y2,linewidth=5,c='#F73668',label='cases')#plot cases
ax2.set_ylim(0,0.4)#set y axis from 0 to 300

plt.yticks(size = 40)
ax2.set_ylabel('Cases (1,000 persons)',fontsize=60)

ax1.xaxis.set_major_locator(ticker.MultipleLocator(1))#present all weeks on x-axis
ax1.yaxis.set_major_locator(ticker.MultipleLocator(0.1))# ensure 2 y-axis have the same interval
ax2.yaxis.set_major_locator(ticker.MultipleLocator(0.04)) # ensure 2 y-axis have the same interval

fig.legend(fontsize=40,loc=(0.75,0.78)) #legend
plt.title('Brunei',fontsize=60)
plt.savefig("F:\movement\GTWR\GTWR1010\FIGURE\Brunei_C.jpg")
```
